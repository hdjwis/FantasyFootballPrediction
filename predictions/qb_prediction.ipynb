```python
import pandas as pd

data_2024 = pd.read_csv("data/NFL_2024_season_data.csv")
data_2024['year'] = 2024
data_2023 = pd.read_csv("data/NFL_2023_season_data.csv")
data_2023['year'] = 2023
data_2022 = pd.read_csv("data/NFL_2022_season_data.csv")
data_2022['year'] = 2022
data_2021 = pd.read_csv("data/NFL_2021_season_data.csv")
data_2021['year'] = 2021
data_2020 = pd.read_csv("data/NFL_2020_season_data.csv")
data_2020['year'] = 2020
```

```python
data = pd.concat([data_2024, data_2023, data_2022, data_2021, data_2020])
data.shape
```

```python
data.head()
```

```python
data.columns
```

```python
data = data.drop(['2:00 PM', '2PP', 'FantPt', 'DKPt', 'FDPt', 'VBD', 'OvRank', '-9999', 'PPR▼'], axis=1)
```

```python
qb_data = data[data['FantPos'] == 'QB']
```

```python
qb_data = qb_data.reset_index()
qb_data.head()
```

```python
qb_data.columns
```

```python
qb_data = qb_data.drop(['index', 'Rk', 'FantPos', 'Tgt', 'Rec', 'Receiving Yds', 'Receiving Y/R', 'Receiving TD'], axis=1)
qb_data.head()
```

```python
qb_data['Player'] = qb_data['Player'].str.replace('+', '', regex=False)
qb_data['Player'] = qb_data['Player'].str.replace('*', '', regex=False)
qb_data.head()
```
