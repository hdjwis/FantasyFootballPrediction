```python
import pandas as pd

data_2024 = pd.read_csv("data/NFL_2024_season_data.csv")
data_2024['year'] = 2024
data_2023 = pd.read_csv("data/NFL_2023_season_data.csv")
data_2023['year'] = 2023
data_2022 = pd.read_csv("data/NFL_2022_season_data.csv")
data_2022['year'] = 2022
data_2021 = pd.read_csv("data/NFL_2021_season_data.csv")
data_2021['year'] = 2021
data_2020 = pd.read_csv("data/NFL_2020_season_data.csv")
data_2020['year'] = 2020
data = pd.concat([data_2024, data_2023, data_2022, data_2021])
data = data.drop(['2:00 PM', '2PP', 'FantPt', 'DKPt', 'FDPt', 'VBD', 'OvRank', '-9999'], axis=1)
te_data = data[data['FantPos'] == 'TE']
te_data = te_data.reset_index()
te_data.columns
```

```python
te_data = te_data.drop(['index', 'Rk', 'FantPos', 'Cmp', 'Passing Att', 'Passing Yds', 'Passing TD', 'Int', 'Rushing Att', 'Rushing Yds', 'Rushing Y/A', 'Rushing TD', 'Total TD'], axis=1)
te_data['Player'] = te_data['Player'].str.replace('+', '', regex=False)
te_data['Player'] = te_data['Player'].str.replace('*', '', regex=False)
te_data.head()
```

```python
te_data[te_data.isna().any(axis=1)]
```

```python
te_data = te_data.fillna(0.0)
te_data[te_data.isna().any(axis=1)]
```

```python
te_data['PPG'] = te_data['PPR']/te_data['G']
```

```python
training = te_data[te_data['year'] != 2024]
predicting = te_data[(te_data['year'] == 2024) | (te_data['year'] == 2023)]
```

```python
final_cols = ['Player', 'Age', 
              '0_ago_G', '0_ago_GS', '0_ago_Tgt', '0_ago_Rec', '0_ago_ReceivingYds', '0_ago_ReceivingY/R', '0_ago_ReceivingTD', '0_ago_Fmb', '0_ago_FL', '0_ago_PPR', 
              '0_ago_PosRank', '0_ago_PPG',
              '1_ago_G', '1_ago_GS', '1_ago_Tgt', '1_ago_Rec', '1_ago_ReceivingYds', '1_ago_ReceivingY/R', '1_ago_ReceivingTD', '1_ago_Fmb', '1_ago_FL', '1_ago_PPR', 
              '1_ago_PosRank', '1_ago_PPG',
              '2_ago_G', '2_ago_GS', '2_ago_Tgt', '2_ago_Rec', '2_ago_ReceivingYds', '2_ago_ReceivingY/R', '2_ago_ReceivingTD', '2_ago_Fmb', '2_ago_FL', '2_ago_PPR', 
              '2_ago_PosRank', '2_ago_PPG'
              ]
te = pd.DataFrame(columns=final_cols, index=[0])
te = te.fillna(0.0)
for player in training["Player"].unique():
    individual = training[training['Player'] == player]
    individual = individual.reset_index()
    data = pd.DataFrame(columns=final_cols, index=[0])
    data['Player'] = player
    data['Age'] = individual.loc[0]['Age']
    for row in individual.itertuples():
        data[str(2023 - row.year) + '_ago_G'] = row.G
        data[str(2023 - row.year) + '_ago_GS'] = row.GS
        data[str(2023 - row.year) + '_ago_Tgt'] = row.Tgt
        data[str(2023 - row.year) + '_ago_Rec'] = row.Rec
        data[str(2023 - row.year) + '_ago_ReceivingYds'] = row._9
        data[str(2023 - row.year) + '_ago_ReceivingY/R'] = row._10
        data[str(2023 - row.year) + '_ago_ReceivingTD'] = row._11
        data[str(2023 - row.year) + '_ago_Fmb'] = row.Fmb
        data[str(2023 - row.year) + '_ago_FL'] = row.FL
        data[str(2023 - row.year) + '_ago_PPR'] = row.PPR
        data[str(2023 - row.year) + '_ago_PosRank'] = row.PosRank
        data[str(2023 - row.year) + '_ago_PPG'] = row.PPG

    te = pd.concat([te, data])

te.head()
```

```python
from sklearn.model_selection import train_test_split

X = te.drop(['0_ago_PPR', 'Player'], axis=1)
y = te[['Player', '0_ago_PPR']]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=0)
```

```python
ls = ['Age', '1_ago_PPR', '2_ago_PPR']
simplified_X = te[ls]
simplified_y = te[['Player', '0_ago_PPR']]
simplified_X_train, simplified_X_test, simplified_y_train, simplified_y_test = train_test_split(simplified_X, simplified_y, train_size=0.75, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=0)
model.fit(simplified_X_train, simplified_y_train['0_ago_PPR'])
model.score(simplified_X_test, simplified_y_test['0_ago_PPR'])
```

```python
y_pred = model.predict(simplified_X_test)
```

```python
y_test = y_test.reset_index()
y_test = y_test.drop(['index'], axis=1)
y_test
```

```python
y_pred = pd.Series(y_pred)
y_pred
```

```python
comp = pd.concat([y_test, y_pred], axis=1)
comp.head()
```

```python
from sklearn.ensemble import GradientBoostingRegressor

xgb_model = GradientBoostingRegressor(random_state=0)
xgb_model.fit(simplified_X_train, simplified_y_train['0_ago_PPR'])
xgb_model.score(simplified_X_test, simplified_y_test['0_ago_PPR'])
```

```python
simplified_y_test = simplified_y_test.reset_index()
simplified_y_test = simplified_y_test.drop(['index'], axis=1)
simplified_y_test
```
