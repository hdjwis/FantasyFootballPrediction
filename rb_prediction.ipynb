```python
import pandas as pd

data_2024 = pd.read_csv("data/NFL_2024_season_data.csv")
data_2024['year'] = 2024
data_2023 = pd.read_csv("data/NFL_2023_season_data.csv")
data_2023['year'] = 2023
data_2022 = pd.read_csv("data/NFL_2022_season_data.csv")
data_2022['year'] = 2022
data_2021 = pd.read_csv("data/NFL_2021_season_data.csv")
data_2021['year'] = 2021
```

```python
data = pd.concat([data_2024, data_2023, data_2022, data_2021])
data = data.drop(['2:00 PM', '2PP', 'FantPt', 'DKPt', 'FDPt', 'VBD', 'OvRank', '-9999'], axis=1)
rb_data = data[data['FantPos'] == 'RB']
rb_data = rb_data.reset_index()
rb_data.columns
```

```python
rb_data = rb_data.drop(['index', 'Rk', 'FantPos', 'Cmp', 'Passing Att', 'Passing Yds', 'Passing TD', 'Int'], axis=1)
rb_data['Player'] = rb_data['Player'].str.replace('+', '', regex=False)
rb_data['Player'] = rb_data['Player'].str.replace('*', '', regex=False)
rb_data.head()
```

```python
rb_data[rb_data.isna().any(axis=1)]
```

```python
rb_data = rb_data.fillna(0.0)
```

```python
rb_data['PPG'] = rb_data['PPR']/rb_data['G']
```

```python
rb_data = rb_data[rb_data['G'] > 5]
```

```python
rb_data.shape
```

```python
training = rb_data[rb_data['year'] != 2024]
predicting = rb_data[(rb_data['year'] == 2024) | (rb_data['year'] == 2023)]
```

```python
rb_data.columns
```

```python
henry = training[training['Player'] == "Derrick Henry"]
henry = henry.reset_index()
for row in henry.itertuples():
    print(row)
    break
```

```python
final_cols = ['Player', 'Age', 
              '0_ago_G', '0_ago_GS', '0_ago_RushingAtt', '0_ago_RushingYds', '0_ago_RushingY/A', '0_ago_RushingTD', '0_ago_Tgt', '0_ago_Rec', '0_ago_ReceivingYds', 
              '0_ago_ReceivingY/R', '0_ago_ReceivingTD', '0_ago_Fmb', '0_ago_FL', '0_ago_TotalTD', '0_ago_PPR', '0_ago_PosRank', '0_ago_PPG', 
              '1_ago_G', '1_ago_GS', '1_ago_RushingAtt', '1_ago_RushingYds', '1_ago_RushingY/A', '1_ago_RushingTD', '1_ago_Tgt', '1_ago_Rec', '1_ago_ReceivingYds', 
              '1_ago_ReceivingY/R', '1_ago_ReceivingTD', '1_ago_Fmb', '1_ago_FL', '1_ago_TotalTD', '1_ago_PPR', '1_ago_PosRank', '1_ago_PPG', 
              '2_ago_G', '2_ago_GS', '2_ago_RushingAtt', '2_ago_RushingYds', '2_ago_RushingY/A', '2_ago_RushingTD', '2_ago_Tgt', '2_ago_Rec', '2_ago_ReceivingYds', 
              '2_ago_ReceivingY/R', '2_ago_ReceivingTD', '2_ago_Fmb', '2_ago_FL', '2_ago_TotalTD', '2_ago_PPR', '2_ago_PosRank', '2_ago_PPG'
            ]

rb = pd.DataFrame(columns=final_cols, index=[0])
rb = rb.fillna(0.0)
for player in training["Player"].unique():
    individual = training[training['Player'] == player]
    individual = individual.reset_index()
    data = pd.DataFrame(columns=final_cols, index=[0])
    data['Player'] = player
    data['Age'] = individual.loc[0]['Age']
    for row in individual.itertuples():
        data[str(2023 - row.year) + '_ago_G'] = row.G
        data[str(2023 - row.year) + '_ago_GS'] = row.GS
        data[str(2023 - row.year) + '_ago_RushingAtt'] = row._7
        data[str(2023 - row.year) + '_ago_RushingYds'] = row._8
        data[str(2023 - row.year) + '_ago_RushingY/A'] = row._9
        data[str(2023 - row.year) + '_ago_RushingTD'] = row._10
        data[str(2023 - row.year) + '_ago_Tgt'] = row.Tgt
        data[str(2023 - row.year) + '_ago_Rec'] = row.Rec
        data[str(2023 - row.year) + '_ago_ReceivingYds'] = row._13
        data[str(2023 - row.year) + '_ago_ReceivingY/R'] = row._14
        data[str(2023 - row.year) + '_ago_ReceivingTD'] = row._15
        data[str(2023 - row.year) + '_ago_Fmb'] = row.Fmb
        data[str(2023 - row.year) + '_ago_FL'] = row.FL
        data[str(2023 - row.year) + '_ago_TotalTD'] = row._18
        data[str(2023 - row.year) + '_ago_PPR'] = row.PPR
        data[str(2023 - row.year) + '_ago_PosRank'] = row.PosRank
        data[str(2023 - row.year) + '_ago_PPG'] = row.PPG

    rb = pd.concat([rb, data])

rb.head()
```
