```python
import pandas as pd

data_2024 = pd.read_csv("data/NFL_2024_season_data.csv")
data_2024['year'] = 2024
data_2023 = pd.read_csv("data/NFL_2023_season_data.csv")
data_2023['year'] = 2023
data_2022 = pd.read_csv("data/NFL_2022_season_data.csv")
data_2022['year'] = 2022
data_2021 = pd.read_csv("data/NFL_2021_season_data.csv")
data_2021['year'] = 2021
data_2020 = pd.read_csv("data/NFL_2020_season_data.csv")
data_2020['year'] = 2020
```

```python
data = pd.concat([data_2024, data_2023, data_2022, data_2021, data_2020])
data.shape
```

```python
data = data.drop(['2:00 PM', '2PP', 'FantPt', 'DKPt', 'FDPt', 'VBD', 'OvRank', '-9999'], axis=1)
```

```python
wr_data = data[data['FantPos'] == 'WR']
```

```python
wr_data = wr_data.reset_index()
wr_data.head()
```

```python
wr_data.columns
```

```python
wr_data = wr_data.drop(['index', 'Rk', 'FantPos', 'Cmp', 'Passing Att', 'Passing Yds', 'Passing TD', 'Int'], axis=1)
```

```python
wr_data['Player'] = wr_data['Player'].str.replace('+', '', regex=False)
wr_data['Player'] = wr_data['Player'].str.replace('*', '', regex=False)
wr_data.head()
```
